```python
from IPython.display import Markdown

markdown_text = '''
# Moderate
1. What are the three most popular categories, by total sales?

2. What are the total sales attributed to customers over the age of 45?

3. How is the data distributed across different invoice dates? Are there any seasonal trends or patterns? (Use a graph for this.)
'''
display(Markdown(markdown_text))
```

```python
file_1['total_revenue']=file_1['quantity']*file_1['price']
file_1

sum_of_category = file_1.groupby('category')["total_revenue"].sum() #groupby asks you by what category do you want certain data to be grouped in 
                                                                    #sum in the end just adds the total

sorted_category = sum_of_category.sort_values(ascending=False)[0:3]

print(sorted_category)

```

```python
new_table = pd.merge(file_1, file_2, how = 'inner', left_on = 'customer_id', right_on ='id')
  #using left_on and right_on alows us to join two different columns from their respective tables

    
age = 45
filtered_table = new_table[new_table['age']> age]
sum_of_45 = filtered_table['total_revenue'].sum()

print(sum_of_45)


```

```python
import matplotlib.pyplot as plt
import time
from datetime import date

new_table['invoice_date'] = pd.to_datetime(new_table['invoice_date'])
new_table['Quarter'] = new_table['invoice_date'].dt.to_period('Q')

quarterly_revenue = new_table.groupby('Quarter').size()

quarterly_revenue.index = quarterly_revenue.index.to_timestamp()

plt.plot(quarterly_revenue.index, quarterly_revenue.values)
plt.xlabel('Quarter')
plt.ylabel('Total Revenue')
plt.title('Revenue by Quarter')
plt.xticks(rotation=45)
plt.show()

print("After the 3rd quarter of 2022, total revenue experienced a sudden drop that was seen till the second quarter of 2023")
```

```python
bins = [0, 9, 19, 29, 39, 49, 59, 69, 79, 89, 99]
labels = ['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-89', '90-99']

new_table['age_range']= pd.cut(new_table['age'], bins=bins, labels=labels, right=False)

pivot_table = pd.pivot_table(new_table, values='total_revenue', index='category', columns = 'age_range', aggfunc='sum')

pivot_table
```
